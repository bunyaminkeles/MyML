# Choosing The Optimal Number of Clusters

## Elbow metod

```python
ssd = []

K = range(2,10)

for k in K:
    model = KMeans(n_clusters =k, random_state=42)
    model.fit(X)
    ssd.append(model.inertia_)
```

```python
# her k değeri için bir inertia değeri alıp SSD içine atıyorum. ssd sum of squares of distance
# bir centroid etrafında bir küme ne kadar yoğun toplanmışsa o kadar iyi
# inertia : ?
# amacım inertia nın min olduğu noktayı bulabilmek
```

```python
# INERTIA : bir kümenin kendi centroidi etrafında ne kadar iyi (ne kadar sıkı şekilde) kümelenmesidir. yani uzaklıkların 
# karelerinin toplamının minimum olduğu değer..  eğer kümelenme çok yayılmış ise iyi bir kümelemeden bahsedemeyiz.
```

```python
# ilk keskin düşüşün durduğu noktanın başlangıcı
# sonradan keskin düşüşler olsa bile ilkini dikkate alıyoruz
```

```python
plt.plot(K, ssd, "bo-")
plt.xlabel("Different k values")
plt.ylabel("inertia-error") 
plt.title("elbow method") 
```

```python
ssd # sum of squared distance
```

```python
pd.Series(ssd).diff()
```

```python
df_diff =pd.DataFrame(-pd.Series(ssd).diff()).rename(index = lambda x : x+1)
df_diff
```

```python
df_diff.plot(kind='bar')
```

```python
from yellowbrick.cluster import KElbowVisualizer

model_ = KMeans(random_state=42)
visualizer = KElbowVisualizer(model_, k=(2,9))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show();
```

## Silhouette analysis

```python
# kendi içinde ne kadar iyi kümelenmiş
# diğer kümelerden ne kadar uzak,
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(X, K_means_model.labels_)
```

```python
range_n_clusters = range(2,9)
for num_clusters in range_n_clusters:
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    # silhouette score
    silhouette_avg = silhouette_score(X, cluster_labels)
    print(f"For n_clusters={num_clusters}, the silhouette score is {silhouette_avg}")
```

```python
# 2 daha iyi siluet skoru vermiş
# ancak, burada inertia skoruna da bakmak gerekli. inertia skoru 2 de çok büyük bir kümelenme olduğunu söylüyor
# amacım centroid etrafında en iyi şekilde yoğunlaşmış kümeleri bulmak
# diğer kümelerden en uzak olacak şekilde kümelenmek
# burada 3 en iyi çözüm
# silüet skoru bana yaptığım kümelenmenin ne kadar kaliteli olduğunu gösterir.
```

```python
from sklearn.cluster import KMeans

from yellowbrick.cluster import SilhouetteVisualizer

model3 = KMeans(n_clusters=3, random_state=42)
visualizer = SilhouetteVisualizer(model3)

visualizer.fit(X)    # Fit the data to the visualizer
visualizer.poof();
```
